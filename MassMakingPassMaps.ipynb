```python
#just tryna copy this https://mplsoccer.readthedocs.io/en/latest/gallery/pitch_plots/plot_arrows.html
import pandas as pd
from mplsoccer import Pitch, FontManager
from matplotlib import rcParams
import matplotlib.pyplot as plt
df = pd.read_csv('/Users/fsa168/Desktop/PythonPractice/SoccerProj1/final_df2.csv', index_col=0)
df
```

```python
ITAsquad = list(df['player'].value_counts().keys())
```

```python
for eachplayer in ITAsquad:
    pitch = Pitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc')
    fig, ax = pitch.draw(figsize=(7,4.8), constrained_layout=True, tight_layout=False)
    fig.set_facecolor('#22312b')
    ax.set_title(f'{eachplayer} Completed Passes', fontsize=20, color='white')

    filt = df['player'] == eachplayer
    player_df = df.loc[filt]
    player_df_slim = player_df[['x_start', 'y_start','x_end','y_end','pass_outcome']]
    player_df_mask_complete = player_df_slim.pass_outcome.isnull()
    player_completed_passes = player_df_slim[player_df_mask_complete]
    
    player_fig = pitch.arrows(player_completed_passes.x_start,player_completed_passes.y_start,player_completed_passes.x_end,player_completed_passes.y_end, width=2, headwidth=5,headlength=5,color='#ad993c', ax=ax, label='completed passes')

    # fig.savefig('/Users/fsa168/Desktop/PythonPractice/SoccerProj1/Figs/' + eachplayer)
```
