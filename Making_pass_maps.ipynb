```python
#just tryna copy this https://mplsoccer.readthedocs.io/en/latest/gallery/pitch_plots/plot_arrows.html
import pandas as pd
from mplsoccer import Pitch, FontManager
from matplotlib import rcParams
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/fsa168/Desktop/PythonPractice/SoccerProj1/final_df2.csv', index_col=0)
df
```

```python
df_slim = df[['x_start', 'y_start','x_end','y_end','pass_outcome']]
df_slim.head()
```

```python
mask_complete = df_slim.pass_outcome.isnull()
```

```python
pitch = Pitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc')
fig, ax = pitch.draw(figsize=(16, 11), constrained_layout=True, tight_layout=False)
fig.set_facecolor('#22312b')

pitch.arrows(df_slim[mask_complete].x_start, df_slim[mask_complete].y_start, df_slim[mask_complete].x_end,df_slim[mask_complete].y_end, width=2, headwidth=10,headlength=10, color='#ad993c', ax=ax, label='completed passes')
```
