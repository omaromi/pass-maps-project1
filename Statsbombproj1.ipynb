```python
ITAdf = df[filt]
ITAgames = list(ITAdf['match_id'])

ITAdf
```

Obviously the final can be interesting. Let's analyze that one.

```python
final_match_id=ITAdf.iloc[1]['match_id']
```

```python
events = sb.events(match_id=final_match_id, split=True, flatten_attrs=False)
#I don't fully understand events at this point but I am following the statsbombpy github repo.
events.keys()
#let me try to find what kinds of events exist i.e. what are the columns
```

Now pause to think.

If I want to create pass maps for each, I think I would need each player's name. And a summary of all their passes. but I'm not sure what the passes DataFrame looks like

```python
columns = list(events['passes'].columns)
columns
```

```python
ITAfilt = ((events['passes']['team'])=='Italy')
ITApasses = events['passes'][ITAfilt]
ITApasses
```

I'm looking for the potential values in each column and see if they are even useful to me.

```python
for each in columns:
    print(ITApasses[each].value_counts())
```

It seems the most useful columns will be:
location i.e. the xy position of the pass start
player i.e. the player committing the pass, and
pass i.e. a dictionary containing the recipient, xy position of the recipient, and data about the style of pass.

I'm basing this off the mplsoccer docs that describes how to make pass plots using arrows here: https://mplsoccer.readthedocs.io/en/latest/gallery/pitch_plots/plot_arrows.html

Time to make a much smaller DataFrame

```python
slimdf = ITApasses[['period','location','player','pass']].copy()
slimdf.reset_index(drop=True,inplace=True)
slimdf
```

I want to get a DataFrame for each player that has columns, x_start, y_start, x_end, y_end, pass success?, pass type

```python
slimdf['location'].hasnans
```

```python
pass_start = pd.DataFrame(slimdf['location'].to_list(),columns=['x_start','y_start'])
pass_start.reset_index(drop=True)
```

```python
slimdf = slimdf.join(pass_start)
```

```python
slimdf
```

```python
print(slimdf['pass'][0].keys())
print(slimdf['pass'][0].get('recipient')['name'])
print(slimdf['pass'][0].get('end_location'))
```

```python
#Had a series of dicts, tried to turn it into a dataframe via https://stackoverflow.com/questions/29681906/python-pandas-dataframe-from-series-of-dict
pass_col_df = slimdf['pass'].apply(pd.Series)
pass_col_df
```

It was at this moment. That I said what the fuck. I had no idea that the pass Series was gonna be a 19 column DataFrame itself

```python
for each in list(pass_col_df.columns):
    print(pass_col_df[each].value_counts())
```

```python
pass_col_df.keys()
```

```python
pass_col_slimmed_df = pass_col_df[['recipient','end_location','type','outcome']].copy()
pass_col_slimmed_df
```

```python
pass_col_slimmed_df['type']
passtypedf = pass_col_slimmed_df['type'].apply(pd.Series)
passtypedf.rename(columns={'name':'pass_type'},inplace=True)
passtypedf = passtypedf['pass_type'].copy()
passtypedf
```

```python
pass_col_slimmed_df['recipient']
recip_name_df = pass_col_slimmed_df['recipient'].apply(pd.Series)
recip_name_df.rename(columns={'name':'recipient_name'},inplace=True)
recip_name_df = recip_name_df['recipient_name'].copy()
recip_name_df
```

```python
pass_col_slimmed_df['outcome']
outcome_df = pass_col_slimmed_df['outcome'].apply(pd.Series)
outcome_df.rename(columns={'name':'pass_outcome'},inplace=True)
outcome_df = outcome_df['pass_outcome'].copy()
outcome_df
```

```python
pass_col_slimmed_df = pass_col_slimmed_df.join([recip_name_df,passtypedf,outcome_df])
pass_col_slimmed_df
```

```python
pass_end = pd.DataFrame(pass_col_slimmed_df['end_location'].to_list(), columns=['x_end', 'y_end'])
pass_col_slimmed_df = pass_col_slimmed_df.join(pass_end)
pass_col_slimmed_df
```

```python
pass_info_df = pass_col_slimmed_df[['recipient_name','x_end','y_end','pass_type','pass_outcome']]
pass_info_df
```

```python
final_df = slimdf.join(pass_info_df)
```

```python
final_df.drop(['location','pass'],axis=1,inplace=True)
```
